```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Feature Engineering: Compute Balance and Debt-to-Income Ratio
monthly_summary["Balance"] = monthly_summary["Credit"] - monthly_summary["Debit"]
monthly_summary["DTI"] = (monthly_summary["Debit"] / monthly_summary["Credit"]).replace([np.inf, -np.inf], np.nan).fillna(0)

# Define Loan Approval Criteria (1 = Approved, 0 = Rejected)
monthly_summary["Loan Approved"] = (monthly_summary["DTI"] < 0.8).astype(int)

from sklearn.utils import shuffle
monthly_summary = shuffle(monthly_summary, random_state=42)


# Define Features and Target
X = monthly_summary[["Credit", "Debit", "Balance", "DTI"]]
y = monthly_summary["Loan Approved"]

# Ensure no infinite or NaN values
X.replace([np.inf, -np.inf], np.nan, inplace=True)
X.fillna(0, inplace=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)


# Train ML Model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Evaluate Model
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print("Model Accuracy:", round(accuracy * 100, 2), "%")

# Predict Loan Approval for the most recent month
latest_data = [X.iloc[-1].values]  # Latest available month
loan_prediction = model.predict(latest_data)

if loan_prediction[0] == 1:
    print("Loan Decision: APPROVED ")
else:
    print("Loan Decision: REJECTED ")
```
